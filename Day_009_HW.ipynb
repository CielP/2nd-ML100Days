```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 負相關

```python
x = np.random.randint(0, 50, 1000)
y = -x + np.random.normal(-10, 10, 1000)
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
